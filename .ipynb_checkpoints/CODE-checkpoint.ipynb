## 1.2 Remove Dodgy Images

Images fetched from the web, despite having appropriate file extensions, can still face issues such as corruption, mislabeling, or incorrect extensions. By implementing this block of code, we can mitigate such issues, ensuring smooth processing within Python.

The `imghdr` module in Python is used to determine the type of image contained in a file or byte stream. It provides functions to identify the format of an image by examining its header or magic number. This can be useful for verifying the integrity of image files, especially when dealing with images from untrusted sources or when the file extension may not accurately reflect the image format.

```python
import cv2
import imghdr
from matplotlib import pyplot as plt
```

```python
path_dir = 'data'
```

```python
os.listdir(path_dir)
```

```python
image_exts = ['jpeg','jpg', 'bmp', 'png']
```

```python
for image_class in os.listdir(path_dir): 
    for image in os.listdir(os.path.join(path_dir, image_class)):
        image_path = os.path.join(path_dir, image_class, image)
        try: 
            img = cv2.imread(image_path)
            tip = imghdr.what(image_path)
            if tip not in image_exts: 
                print('Image not in ext list {}'.format(image_path))
                os.remove(image_path)
        except Exception as e: 
            print('Issue with image {}'.format(image_path))
            # os.remove(image_path)
```

# 2. Preprocess the Data

### 2.1 Scale Data

This line of code applies a normalization transformation to the input data x by dividing it by 255, effectively scaling pixel values to the range [0, 1]. It preserves the label data y unchanged. The .next() function retrieves the next batch of data from the iterator, allowing inspection of the transformed data.

```python
data = data.map(lambda x,y: (x/255, y))
```

```python
scaled_data = data.as_numpy_iterator()
batch = scaled_data.next()
batch
```

```python
batch[0].max()
```

```python
fig, ax = plt.subplots(ncols=4, figsize=(20,20))
for idx, img in enumerate(batch[0][:4]):
    ax[idx].imshow(img)
    ax[idx].title.set_text(batch[1][idx])
```

### 2.2 Split Data

```python
len(data)
```

```python
train_size = int(len(data)*.7)    # 70%  of the data is in Training set
val_size = int(len(data)*.15)      # 15% of the data is for validation
test_size = int(len(data)*.15)     # 15% of the data is for testing
```

```python
train_size
```

```python
val_size
```

```python
test_size
```

```python
train = data.take(train_size)
val = data.skip(train_size).take(val_size)
test = data.skip(train_size+val_size).take(test_size)
```

# 3. Building a Model

```python
train
```

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
```

```python
model = Sequential()
model.add(Conv2D(16, (3,3), 1, activation='relu', input_shape=(256,256,3)))
model.add(MaxPooling2D())
model.add(Conv2D(32, (3,3), 1, activation='relu'))
model.add(MaxPooling2D())
model.add(Conv2D(16, (3,3), 1, activation='relu'))
model.add(MaxPooling2D())
model.add(Flatten())
model.add(Dense(256, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile('adam', loss=tf.losses.BinaryCrossentropy(), metrics=['accuracy'])
```

```python
model.summary()
```

Train

```python
logdir='logs'
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=logdir)
hist = model.fit(train, epochs=15, validation_data=val, callbacks=[tensorboard_callback])
```

```python
fig = plt.figure()
plt.plot(hist.history['loss'], color='teal', label='loss')
plt.plot(hist.history['val_loss'], color='orange', label='val_loss')
fig.suptitle('Loss', fontsize=20)
plt.legend(loc="upper left")
plt.show()
```

```python
fig = plt.figure()
plt.plot(hist.history['accuracy'], color='teal', label='accuracy')
plt.plot(hist.history['val_accuracy'], color='orange', label='val_accuracy')
fig.suptitle('Accuracy', fontsize=20)
plt.legend(loc="upper left")
plt.show()
```

## 4. Evaluate the Model

```python
from tensorflow.keras.metrics import Precision, Recall, BinaryAccuracy
```

```python
pre = Precision()
re = Recall()
acc = BinaryAccuracy()
```

```python
for batch in test.as_numpy_iterator(): 
    X, y = batch
    yhat = model.predict(X)
    pre.update_state(y, yhat)
    re.update_state(y, yhat)
    acc.update_state(y, yhat)
```

```python
print(f'Precision: {pre.result().numpy()}, Recall:{re.result().numpy()}, Accuracy:{acc.result().numpy()}')
```

## 5. Testing the Model

```python
import cv2
```

```python
img = cv2.imread('happytest.jpeg')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
resized_img = tf.image.resize(img, (256, 256))
resized_img = cv2.cvtColor(resized_img.numpy().astype(np.uint8), cv2.COLOR_BGR2RGB)  # Convert to RGB
plt.imshow(resized_img)
plt.show()
```

```python
yhat = model.predict(np.expand_dims(resized_img/255, 0))
yhat

if yhat > 0.5: 
    print(f'Predicted class is Sad')
else:
    print(f'Predicted class is Happy')
```
